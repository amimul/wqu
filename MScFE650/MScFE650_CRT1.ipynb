### 1.1. Solving the system:
$$ x_1 = 1.0 $$
$$ x_2 = 0.5 $$
$$ x_3 = 0.7 $$

#### 1.1.1. Write the system in form of $ Ax = b $: 
We have: $ A = \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}$ , $ x = \begin{bmatrix} x1 \\ x2 \\ x3 \end{bmatrix} $, and $ b = \begin{bmatrix} 1.0 \\ 0.5 \\ 0.7 \end{bmatrix} $

Thus the system will be:
$ \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix} . \begin{bmatrix} x1 \\ x2 \\ x3 \end{bmatrix} = \begin{bmatrix} 1.0 \\ 0.5 \\ 0.7 \end{bmatrix} $

#### 1.1.2. Write down the normal equations, $ A^T * Ax = A^T * b $: 

$ \begin{bmatrix} 1 && 1 && 1 \end{bmatrix} . \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix} . \begin{bmatrix} x1 \\ x2 \\ x3 \end{bmatrix} = \begin{bmatrix} 1 && 1 && 1 \end{bmatrix} . \begin{bmatrix} 1.0 \\ 0.5 \\ 0.7 \end{bmatrix} $

#### 1.1.3. Solve the normal equation above:

Since the unknown is x, we will solve this to get x, which is
$$ x = (A^T . A)^{-1} . (A^T . b) $$
with: 
$ A = \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}$, $ A^T = \begin{bmatrix} 1 && 1 && 1 \end{bmatrix} $ , $ x = \begin{bmatrix} x1 \\ x2 \\ x3 \end{bmatrix} $, and $ b = \begin{bmatrix} 1.0 \\ 0.5 \\ 0.7 \end{bmatrix} $

### 1.2. Given pair of datas (x, y)

#### 1.2.1. Write this as a linear system where the unknowns are <i> m </i> and <i> b </i>:

Let's write this as matrix form:

$$ mx_n + b = y_n $$ 
$$ => X.w = Y $$

where w is the unknown, $ w = \begin{bmatrix} m \\ b \end{bmatrix} $, $ X = \begin{bmatrix} 1 && x_1 \\ ... && ... \\ 1 && x_n \end{bmatrix} $, $ Y = \begin{bmatrix} 1 && y_1 \\ ... && ... \\ 1 && y_n \end{bmatrix} $

$$ => \begin{bmatrix} 1 && x_1 \\ ... && ... \\ 1 && x_n \end{bmatrix}.\begin{bmatrix} m \\ b \end{bmatrix} = \begin{bmatrix} 1 && y_1 \\ ... && ... \\ 1 && y_n \end{bmatrix} $$ 

#### 1.2.2. Write down and solve the normal equations for <i> m </i> and <i> b </i>:

Normal Equation: 

$$ X^T.Xw = X^T.Y $$
$$ => w = (X^T . X)^{-1} . (X^T . Y) $$

where:

$ X^T = \begin{bmatrix} 1 && ... && 1 \\ x_1 && ... && x_n \end{bmatrix} $

### 1.3. Optimization problem:
We minimize the mean squared error by:
$$ E = \frac{1}{N} \sum_{n=1}^N (mx_n + b - yn)^2 $$

#### 1.3.1. Write down the equation $ \frac{\partial E}{\partial m} = 0 $ and $ \frac{\partial E}{\partial b} = 0 $:

$\frac{\partial E}{\partial m} = \frac{2}{N}\sum_{i=1}^N (mx_n +b -y_n) x_n = \frac{2}{N}\sum_{i=1}^N (mx_n^2 +bx_n -y_n x_n)$


$\frac{\partial E}{\partial b} = \frac{2}{N}\sum_{i=1}^N (mx_n +b -y_n)$

#### 1.3.2. Solve the system above
Since $ \frac{\partial E}{\partial m} = 0 $ and $\frac{\partial E}{\partial b} = 0 $ , then $ \sum_{i=1}^N (mx_n^2 +bx_n -y_n x_n) = 0 $, and  $ \sum_{i=1}^N (mx_n +b -y_n) = 0 $

Exactly like:

$ w = (X^T . X)^{-1} . (X^T . Y) $

where:

$ X^T = \begin{bmatrix} 1 && ... && 1 \\ x_1 && ... && x_n \end{bmatrix} $, 
$ w = \begin{bmatrix} m \\ b \end{bmatrix} $, $ X = \begin{bmatrix} 1 && x_1 \\ ... && ... \\ 1 && x_n \end{bmatrix} $, $ Y = \begin{bmatrix} 1 && y_1 \\ ... && ... \\ 1 && y_n \end{bmatrix} $

```python
#%%
import numpy as np # import the numerical linear algebra model
import matplotlib.pyplot as plt
#%%
x_train = np.linspace(0,1,100)
y_train = 0.2*x_train + 1 + 0.01 * np.random.randn(x_train.shape[0])
x_train_1 = np.array([x_train.tolist()]).T
plt.plot(x_train, y_train, 'r.')
plt.show()
```

```python
# In here, we know that we need to do a manual linear regression, thus we will
# use the following equation
# Xw = Y
# Using normal equation like the equation above, with transpose and invert
# Matrix
# Building the X array
one = np.ones((x_train_1.shape[0], 1))
X = np.concatenate((one, x_train_1), axis = 1)
Y = y_train
# Calculate w = [[m, b]]
# left = (X^T . X)
left = np.dot(X.T, X)
# right = (X^T . Y)
right = np.dot(X.T, Y)
# w = invert(left) . right
w = np.dot(np.linalg.pinv(left), right)
```

```python
m = w[0]
b = w[1]
```
